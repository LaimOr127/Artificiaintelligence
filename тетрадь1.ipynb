```python
import numpy as np 
import matplotlib.pyplot as plt 
 
# задаем размер массива 
n = 100 
 
# генерируем массив случайных значений из интервала (0; 1) 
data = np.random.uniform(0, 1, n) 
 
# рассчитываем среднее значение 
mean = np.mean(data) 
 
# рассчитываем медиану 
median = np.median(data) 
 
# выводим результаты 
print("Среднее значение: ", mean) 
print("Медиана: ", median) 
 
# строим точечную диаграмму рассеяния 
plt.scatter(range(n), data) 
plt.title("Точечная диаграмма рассеяния") 
plt.xlabel("Индекс элемента") 
plt.ylabel("Значение элемента") 
plt.show()
```

```python
import math
print("Ноль в качестве знака операции"
      "\nзавершит работу программы")
while True:
    s = input(" Знак ('+', '-', '*', '/', 'e**a+b','sin(a+b)','cos(a+b)','a**b'): ")
    if s == '0': # завершение программы
        break
    if s in ('+', '-', '*', '/', 'e**a+b','sin(a+b)','cos(a+b)','a**b'):
        a = float(input("a="))
        b = float(input("b="))
        if s == '+':
            print("%.2f" % (a+b)) # слодение
        elif s == '-':
            print("%.2f" % (a-b)) # вычетание
        elif s == 'e**a+b':
            print((2,71828)**(a+b)) # степень со слож.
        elif s == 'sin(a+b)':
            print(math.sin(a++b)) # sin со слож.
        elif s == 'cos(a++b)':
            print(math.cos(a+b)) # cos со слож.
        elif s == 'a**b':
            print(a**b) # степень
        elif s == '*':
            print("%.2f" % (a*b)) # умножение
        elif s == '/':
            if b != 0:
                print("%.2f" % (a/b)) # деление
            else:
                print("Деление на ноль!")
    else:
        print("Неверный знак операции!")
```
