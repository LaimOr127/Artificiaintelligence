```python
# Задаем размер матрицы
n = 5
m = 5

# Создаем матрицу с нулями
matrix = [[0 for j in range(m)] for i in range(n)]

# Заполняем границы матрицы единицами
for i in range(n):
    for j in range(m):
        if i == 0 or j == 0 or i == n-1 or j == m-1:
            matrix[i][j] = 1

# Выводим матрицу
for i in range(n):
    for j in range(m):
        print(matrix[i][j], end=' ')
    print()
```

Библиотека pandas предоставляет две ключевые структуры данных:
Series и DataFrame.

Series – это одномерная структура данных, ее можно представить, как
таблицу с одной строкой. С Series можно работать как с обычным массивом
(обращаться по номеру индекса), и как с ассоциированным массивом, когда
можно использовать ключ для доступа к элементам данных.

DataFrame – это двумерная структура. Идейно она очень похожа на
обычную таблицу, что выражается в способе ее создания и работе с ее
элементами

2.3.1

```python
import pandas as pd
import numpy as np

# Создаем две Series (точки)
a = pd.Series([1, 2, 3])
b = pd.Series([4, 5, 6])

# Вычисляем квадраты разностей между соответствующими элементами a и b
squares = (a - b) ** 2

# Суммируем полученные квадраты
sum_of_squares = np.sum(squares)

# Извлекаем квадратный корень из полученной суммы
result = np.sqrt(sum_of_squares)
print("Евклидово расстояние между a и b:", result)
```

```python
import pandas as pd

# Загружаем фрейм данных из CSV файла
df = pd.read_csv('iris.csv')
# Вычисляем минимальное и максимальное значение для первого признака
x_min = df["sepal_length_cm"].min()
x_max = df["sepal_length_cm"].max()

# Применяем минимаксное преобразование к первому признаку
df["sepal_length_cm_norm"] = (df["sepal_length_cm"] - x_min) / (x_max - x_min)

# Вычисляем среднее значение и стандартное отклонение для второго признака
mean = df["sepal_width_cm"].mean()
std = df["sepal_width_cm"].std()

# Применяем z-масштабирование ко второму признаку
df["sepal_width_cm_norm"] = (df["sepal_width_cm"] - mean) / std

# Выводим обновленный фрейм данных
print(df.head())
```
